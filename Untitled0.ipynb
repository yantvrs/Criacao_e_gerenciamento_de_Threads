```python
import numpy as np
import cv2
import threading
import matplotlib.pyplot as plt

# Função para calcular Gx
def calculate_Gx():
    global Gx
    for i in range(1, M - 1):
        for j in range(1, N - 1):
            Gx[i, j] = abs(I[i + 1, j - 1] + I[i + 1, j] + I[i + 1, j + 1] - (I[i - 1, j - 1] + I[i - 1, j] + I[i - 1, j + 1]))
            Gx[i, j] = max(0, min(255, Gx[i, j]))

# Função para calcular Gy
def calculate_Gy():
    global Gy
    for i in range(1, M - 1):
        for j in range(1, N - 1):
            Gy[i, j] = abs(I[i - 1, j + 1] + I[i, j + 1] + I[i + 1, j + 1] - (I[i - 1, j - 1] + I[i, j - 1] + I[i + 1, j - 1]))
            Gy[i, j] = max(0, min(255, Gy[i, j]))

# Função para calcular G a partir de Gx e Gy com ajuste de limiar
def calculate_G():
    global G
    for i in range(M):
        for j in range(N):
            G[i, j] = Gx[i, j] + Gy[i, j]
            G[i, j] = 0 if G[i, j] < 100 else 255  # Ajuste o limiar (100 neste exemplo) conforme necessário

# Carregando a imagem em nível de cinza (substitua 'image.jpg' pelo nome do arquivo da imagem)
I = cv2.imread('/content/coins.png', cv2.IMREAD_GRAYSCALE)

# Obtendo as dimensões da imagem
M, N = I.shape

# Inicializando as imagens Gx, Gy e G
Gx = np.zeros((M, N), dtype=np.uint8)
Gy = np.zeros((M, N), dtype=np.uint8)
G = np.zeros((M, N), dtype=np.uint8)

# Criando as threads para calcular Gx e Gy
thread_Gx = threading.Thread(target=calculate_Gx)
thread_Gy = threading.Thread(target=calculate_Gy)

# Iniciando as threads
thread_Gx.start()
thread_Gy.start()

# Aguardando a conclusão das threads
thread_Gx.join()
thread_Gy.join()

# Calculando a imagem de saída G
calculate_G()

# Exibindo as imagens
plt.figure(figsize=(10, 10))

# Imagem de entrada
plt.subplot(2, 2, 1)
plt.imshow(I, cmap='gray')
plt.title('Imagem de Entrada')
plt.axis('off')

# Imagem Gx
plt.subplot(2, 2, 2)
plt.imshow(Gx, cmap='gray')
plt.title('Gx')
plt.axis('off')

# Imagem Gy
plt.subplot(2, 2, 3)
plt.imshow(Gy, cmap='gray')
plt.title('Gy')
plt.axis('off')

# Imagem de saída G
plt.subplot(2, 2, 4)
plt.imshow(G, cmap='gray')
plt.title('Imagem de Saída G')
plt.axis('off')

plt.show()
```
